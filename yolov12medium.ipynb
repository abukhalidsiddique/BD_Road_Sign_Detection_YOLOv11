```python
import yaml

data_yaml_path = "/content/BRSDD/data.yaml"

names = [
'Crossroads', 'Emergency Stopping', 'Emergency Stopping 250m', 'Give Way',
'Height Limit 5-7m', 'Hospital Ahead', 'Junction Ahead', 'Mosque Ahead',
'No Overtaking', 'No Pedestrians', 'No Vehicle Entry', 'Pedestrians Crossing',
'Petrol Pump Ahead', 'School Ahead', 'Sharp Left Turn', 'Sharp Right Turn',
'Side Road On Left', 'Side Road On Right', 'Speed Breaker',
'Speed Limit 20 km', 'Speed Limit 40Km', 'Speed Limit 80Km',
'Tolls 1 km Ahead', 'Tolls Ahead', 'Traffic Merges From Left',
'Traffic Merges From Right', 'Truck Lane', 'U Turn', 'Underpass Ahead'
]

data = {
    'train': '/content/BRSDD/train/images',
    'val': '/content/BRSDD/valid/images',
    'test': '/content/BRSDD/test/images',
    'nc': len(names),
    'names': names
}

with open(data_yaml_path, 'w') as f:
    yaml.dump(data, f)

print("✅ data.yaml UPDATED")
```

```python
from ultralytics import YOLO

model = YOLO('yolov12m.pt')   # pretrained or use yolov12m.yaml for training from scratch

model.train(
    data="/content/BRSDD/data.yaml",
    epochs=20,
    imgsz=640,
    batch=16,
    device=0,
     lr0=0.01,
    name="yolov12m_brsdd",
    patience=30
)
```

```python
model = YOLO('/content/runs/detect/yolov12m_brsdd2/weights/best.pt')
model.val(data="/content/BRSDD/data.yaml", imgsz=640)
```

```python
model.predict(source="/content/BRSDD/test/images", imgsz=640, save=True)
```
